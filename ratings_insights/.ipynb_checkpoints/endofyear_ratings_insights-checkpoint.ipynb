```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import init_notebook_mode,iplot,plot
init_notebook_mode(connected=True)
import seaborn as sns
%matplotlib inline
import plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
latest_ratings = pd.read_csv("latest_merlin_ratings.csv", index_col='id')
latest_ratings.shape
```

```python
latest_ratings.dtypes
```

```python
latest_ratings['business_type'] = latest_ratings['business_type'].astype(str)
latest_ratings['staff_pick'] = latest_ratings['staff_pick'].astype(str)
```

```python
ratings_range_text = []
for rating in latest_ratings['overall_rating']:
    
    if rating < 2:
        ratings_range_text.append('1-2')
    if 2 <= rating <= 3:
        ratings_range_text.append('2-3')
    if 3 <= rating <= 4:
        ratings_range_text.append('3-4')
    if rating >= 4:
        ratings_range_text.append('4-5')

latest_ratings['ratings_range_text'] = ratings_range_text

latest_ratings.head()
```

```python
business_type_dict = {
    '0': '',
    '65': 'Life Style',
    '66': 'High Growth',
    '277': 'Growth'
}

staff_pick_dict = {
    '0': '',
    '247': 'Top Deal',
    '248': 'Deal to Watch',
    '249': 'Underweight',
    '250': 'Pass',
    '6163': 'Neutral'
}

latest_ratings['business_type'] = latest_ratings['business_type'].map(business_type_dict)
latest_ratings['staff_pick'] = latest_ratings['staff_pick'].map(staff_pick_dict)
```

```python
latest_ratings.dtypes
```

```python
new_df = latest_ratings[['name', 'security_type', 'state_name', 'overall_rating', 'ratings_range_text', 'business_type', 'staff_pick', 'annual_revenue', 'price_rating', 'market_rating', 'team_rating', 'performance_rating', 'differentiation_rating']]
new_df.head()
```

```python
states_pvt_table_deal_count = new_df.pivot_table(values=['overall_rating'], index=['state_name'], aggfunc=[np.mean, lambda x: len(x.unique())])
states_pvt_table_deal_count.columns = ['avg_overall_rating', '# of deals']

flattened_states_deal_count = pd.DataFrame(states_pvt_table_deal_count.to_records())
flattened_states_deal_count
```

```python
states_pvt_table = new_df.pivot_table(values=['overall_rating'], index=['state_name'])
flattened_states_pvt_table = pd.DataFrame(states_pvt_table.to_records())

pd.options.display.float_format = "{:,.2f}".format
flattened_states_pvt_table
```

```python
data = dict(type='choropleth',
            colorscale = 'Spectral',
            locations = flattened_states_pvt_table['state_name'],
            z = flattened_states_pvt_table['overall_rating'],
            locationmode = 'USA-states',
            text = flattened_states_pvt_table['overall_rating'],
            colorbar = {'title':"Rating"}
            )
```

```python
layout = dict(title = 'State\'s Average Rating',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)')
             )

choromap = go.Figure(data = [data],layout = layout)

plot(choromap,validate=False)
```
